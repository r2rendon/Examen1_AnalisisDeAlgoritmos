**Analisis de complejidad**
---

**O((E log(V))**

*Siendo V el vertice y E la arista*

La complejidad dada en notacion Big O es esa ya que cada arista es insertada a una cola de prioridad una vez y la insercion en dicha cola toma tiempo logaritmico. El algoritmo busca en base el peso de las aristas hacer una cola ordenada por el peso mismo. La arista va siendo multiplicada por *log(V)* ya que el indice del ciclo en el algoritmo es dado a partir del vertice a medida que se va recorriendo el grafo tomando en consideracion los pesos y su prioridad mencionada anteriormente.

---

**Analisis cualitativo**
---
Este algoritmo es uno que busca recorrer un grafo a partir de los pesos que tienen las aristas desde el nodo raiz. El nodo raiz puede ser el ultimo o el primero. Estos pesos dan la prioridad a la cola y tambien dan la prioridad de recorrido para el algoritmo (La prioridad es dada de menor peso a mayor peso). Cabe destacar que si un vertice ya se encuentra en la cola, no se inserta en la cola de prioridad. Es interesante como los pesos en las aristas toman un gran rol ya que es el factor determinante del recorrido de la estructura y de la cola.
